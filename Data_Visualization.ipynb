```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Desktop/clean_hotel_booking.csv')
pd.set_option('display.max_columns',None)
data
```

```python
data.head()
```

```python
sns.countplot(x=data['hotel'])
plt.title("Hotels",fontsize=20)
```

```python
data['is_canceled'].value_counts().plot(kind='pie',autopct='%0.0f%%')
plt.title('Cancelation')
plt.ylabel('')
```

```python
sns.countplot(x=data['adults'])
```

```python
sns.countplot(x=data['children'])
```

```python
sns.countplot(x=data['babies'])
```

```python
people=['adults','children','babies']
fig, ax=plt.subplots(1,3,figsize=(25,7))
#ax[0].plot([1,2,3],[5,8,7])
for i,p in enumerate(people):
    fig.sca(ax[i])
    sns.countplot(data=data,x=p)
    plt.title(f'{p.title()}',fontsize=30)
fig.suptitle('People',fontsize=30,y=1,color='brown')
```

```python
nights=['stays_in_weekend_nights','stays_in_week_nights']
fig, ax=plt.subplots(1,2,figsize=(30,8))
for i,n in enumerate(nights):
    fig.sca(ax[i])
    sns.countplot(data=data,x=n)
    plt.title(f'{n.title()}',fontsize=30)
```

```python
sns.countplot(data=data,x=data['country'],order=data['country'].value_counts().index[:10])
```

```python
def show_plot(col,hue=None):
    plt.figure(figsize=(25,7))
    sns.countplot(data=data,x=col)
    plt.title(f'{col.title()}')
```

```python
show_plot('is_canceled')
```

```python
sns.countplot(x=data['hotel'],hue=data['is_canceled'])
```
