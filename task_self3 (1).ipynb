# Cleaning Data

* Dealing with Missing Values

```python
data.replace(['?'],np.nan,inplace=True)
```

```python
data
```

```python
data.isna().sum()
```

```python
data['num-of-doors'].fillna(method='bfill',inplace=True)
data
```

* Errors Type of Columns

```python
data['normalized-losses'].fillna(0,inplace=True)
data['bore'].fillna(0,inplace=True)
data['stroke'].fillna(0,inplace=True)
data['horsepower'].fillna(0,inplace=True)
data['peak-rpm'].fillna(0,inplace=True)
data['price'].fillna(0,inplace=True)

data['normalized-losses']=data['normalized-losses'].astype(object).astype(int)
data['horsepower']=data['stroke'].astype(object).astype(float)
data['peak-rpm']=data['stroke'].astype(object).astype(float)
data['price']=data['stroke'].astype(object).astype(float)
data['stroke']=data['stroke'].astype(object).astype(float)
data['bore']=data['bore'].astype(object).astype(float)
```

```python
data.isna().sum()
```

```python
data.info()
```

```python
df=data['normalized-losses'].mean()
data['normalized-losses'].replace(0,df,inplace=True)
df2=data['bore'].mean()
data['bore'].replace(0,df2,inplace=True)
df3=data['stroke'].mean()
data['stroke'].replace(0,df3,inplace=True)
df4=data['horsepower'].mean()
data['horsepower'].replace(0,df4,inplace=True)
df5=data['peak-rpm'].mean()
data['peak-rpm'].replace(0,df5,inplace=True)
df6=data['price'].mean()
data['price'].replace(0,df6,inplace=True)
data
```
